```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def get_loc(loc1, loc2, table):
    return [loc1 == el[:len(loc1)] for el in table.index.values], \
           [loc2 == el[:len(loc2)] for el in table.columns]
```

```python
table=pd.read_csv('/Users/alexey/Dropbox/coding/other/hi-c/data/GSM2176967_GB180_norm_int.csv', index_col=0)
table.head()
```

```python
plt.figure(figsize=(15, 15))
sns.heatmap(table.values.astype(bool), vmin=0, vmax=1, xticklabels=False, yticklabels=False, square=True, 
            cmap="YlGnBu")
#По всем хромосомам, 
#как раз видны прямые линии, экстрахромосомные онкогены
```

```python
ch3ch13 = get_loc('chr3', 'chr13', table)
```

```python
plt.figure(figsize=(15, 15))
sns.heatmap(table.iloc[ch3ch13], vmin=0, vmax=0.02, xticklabels=False, 
            yticklabels=False, cmap="YlGnBu")
```

```python
plt.figure(figsize=(15, 15))
sns.heatmap(table.iloc[ch3ch13].values.astype(bool), vmin=0, vmax=1, xticklabels=False, 
            yticklabels=False, cmap="YlGnBu")
```
